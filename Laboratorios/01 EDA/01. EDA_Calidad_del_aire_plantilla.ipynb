# **Base teórica**



**Definición**

El Análisis Exploratorio de Datos (EDA, por sus siglas en inglés: Exploratory Data Analysis) es una etapa fundamental en cualquier proyecto de análisis de datos o ciencia de datos. Su objetivo principal es entender la **estructura, patrones, tendencias, relaciones y anomalía**s en un conjunto de datos antes de aplicar modelos predictivos o tomar decisiones.

**Utilidad**
- Comprender la naturaleza de los datos.
- Detectar valores atípicos (outliers).
- Identificar valores faltantes.
- Verificar supuestos estadísticos.
- Descubrir relaciones entre variables.
- Guiar la selección de técnicas estadísticas o modelos de machine learning.

**Técnicas y herramientas**
1. **Resumen estadístico**
 - Media, mediana, moda
 - Mínimo, máximo
 - Desviación estándar, varianza
 - Cuartiles

2. **Visualización de datos**
 - Histogramas (distribución de una variable)
 - Diagramas de caja (boxplots) para outliers
 - Gráficos de dispersión (relaciones entre variables)
 - Mapas de calor (correlación entre variables)
 - Gráficos de líneas o series temporales (cuando hay datos por fecha)

3. **Análisis de valores faltantes**
 - Detección de NaN o null
 - Métodos para tratarlos (eliminación, imputación, etc.)

4. **Análisis de correlación**
 - Coeficiente de correlación de Pearson o Spearman
 - Identificar relaciones lineales o no lineales



**Pasos a seguir en el EDA**

(https://datos.gob.es/sites/default/files/doc/file/infografia-analisis-exploratorio-datos_0.pdf)
1. **Análisis Descriptivo**
 - **¿Qué es?** Síntesis de la información que proporciona el conjunto de datos, extrayendo sus características más representativas.
  - **¿Por qué es necesario?** Permite conocer los tipos de datos, descubrir patrones y preparar los datos para futuros análisis.
   - **Tratamiento**: Aplicar funciones de estadística descriptiva para explorar la estructura del conjunto de datos, examinar los datos y las variables que presenta.  
2. **Re-ajuste de los tipos de variables**
 - **¿Qué es?** Verificar que las variables se han almacenado con el tipo de valor correspondiente.
  - **¿Por qué es necesario?** Una mala codificación de las variables puede influir negativamente en la agrupación de los datos o los resultados de los análisis.
  - **Tratamiento**: Aplicar la codificación apropiada para cada una de las variables.
3. **Detección y tratamiento de datos ausentes**
  - **¿Qué es?** Identificar la falta de algunos de los datos en la variable.
  - **¿Por qué es necesario?** Los datos ausentes pueden generar problemas a la hora de aplicar técnicas de machine learning, elaborar modelos predictivos, realizar análisis estadísticos o generar representaciones gráficas.
  - **Tratamiento**: Existen varias maneras de tratar los valores ausentes, como por ejemplo sustituirlos por la media o la mediana, o completar los valores faltantes con el valor anterior o posterior de la columna.
4. **Detección y tratamiento de datos atípicos**
 - **¿Qué es?** Identificar datos con valores significativamente distintos a los que presenta la variable.
  -**¿Por qué es necesario su tratamiento?** Pueden modificar los resultados y restar potencia a los análisis estadísticos o técnicas de machine learning aplicadas.
  - **Tratamiento**: Disminuir su influencia en análisis posteriores o, en casos muy extremos, eliminarlos del conjunto de datos.
5. **Análisis de corrrelación de variables**
 - **¿Qué es?** Analizar la relación entre dos o más variables.
  - **¿Por qué es necesario?** Entre otras razones, para descartar posibles variables que aporten información redundante en el conjunto de datos, ocasionando ruido en los análisis.
   - **Tratamiento**: Calcular los coeficientes de correlación para las variables para detectar coeficientes cercanos a 1 ó -1.

## **Descripción del dataset**

El dataset a **utilizar** es un registro de datos diarios de concentración de contaminantes registrados en las estaciones de control de la calidad del aire de la Junta de Castilla y León. (link: https://datos.gob.es/es/catalogo/a07002862-calidad-del-aire-por-dias1)

El dataset contiene las siguientes características (atributos):
- **Fecha**: Fecha del registro

**Contaminantes atmosféricos:** Cada uno de los siguientes atributos representa la concentración de un contaminante específico en el aire:
- **CO (mg/m³)**: Monóxido de carbono. Se expresa en miligramos por metro cúbico.
- **NO (µg/m³)**: Óxido nítrico. Se expresa en microgramos por metro cúbico.
- **NO₂ (µg/m³)**: Dióxido de nitrógeno. Se expresa en microgramos por metro cúbico.
- **O₃ (µg/m³)**: Ozono troposférico. Se expresa en microgramos por metro cúbico.
- **PM10 (µg/m³)**: Partículas en suspensión con diámetro menor o igual a 10 micras. Expresado en microgramos por metro cúbico.
- **PM2.5 (µg/m³)**: Partículas en suspensión con diámetro menor o igual a 2.5 micras. También en microgramos por metro cúbico.
- **SO₂ (µg/m³)**: Dióxido de azufre. Se expresa en microgramos por metro cúbico.

**Ubicación geográfica**
- **Provincia**: Provincia de Castilla y León donde se encuentra la estación de medición.
- **Estación**: Nombre o identificación de la estación de control de calidad del aire.
- **Latitud**: Coordenada geográfica de latitud de la estación (en grados decimales).
- **Longitud**: Coordenada geográfica de longitud de la estación (en grados decimales).
- **Posición**: Coordenadas completas combinadas en un solo campo (latitud, longitud).


## 0. **Carga de datos y vista general**

```python
# Cargar las librerías necesarias
import pandas as pd
import os

# Cargar los datos en un DataFrame
archivo = 'calidad-del-aire-datos-historicos-diarios.csv'
calidad_aire = pd.read_csv(archivo, sep=';')

print(f"Archivo cargado: {archivo}")
print(f"Dimensiones: {calidad_aire.shape}")
```

```python
# Mostrar la estructura del DataFrame
print("Estructura del Dataframe")
print("=" * 50)
print(calidad_aire.info())

# Mostrar un resumen estadístico de las variables numéricas
print("\nResumen estadístico de las variables numéricas")
print("=" * 50)
print(calidad_aire.describe())
```

## 1. **Distribución de las variables numéricas**

```python
import matplotlib.pyplot as plt
import numpy as np

# Seleccionar solo las columnas numéricas
columnas_num = calidad_aire.select_dtypes(include=np.number).columns.tolist()
# Quitamos coordenadas para enfocarnos en contaminantes
columnas_contaminantes = [c for c in columnas_num if c not in ('Latitud', 'Longitud')]

# Calcular el número de filas y columnas para el subplot
n = len(columnas_contaminantes)
ncols = 3
nrows = int(np.ceil(n / ncols))

# Crear la figura y los subplots
fig, axes = plt.subplots(nrows, ncols, figsize=(14, 3.5 * nrows))

# Aplanar el array de ejes en caso de que sea 2D
axes = axes.flatten()

# Crear histogramas para cada variable numérica
for i, col in enumerate(columnas_contaminantes):
    axes[i].hist(calidad_aire[col].dropna(), bins=50, color='steelblue', edgecolor='black')
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frecuencia')

# Ocultar subplots vacíos si los hay
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

## 2. **Ajuste de las variables categóricas**

```python
# Ajustar el tipo de la variable Fecha
calidad_aire['Fecha'] = pd.to_datetime(calidad_aire['Fecha'], format='%Y-%m-%d', errors='coerce')

# Ajustar el tipo de la variable Provincia
calidad_aire['Provincia'] = calidad_aire['Provincia'].astype('category')

# Ajustar el tipo de la variable Estación
calidad_aire['Estación'] = calidad_aire['Estación'].astype('category')

# Verificación
print(calidad_aire.dtypes)
```

## 3. **Análisis de datos ausentes (faltantes)**

```python
# Devuelve un DataFrame booleano
mask_nan = calidad_aire.isnull()
print("DataFrame booleano (primeras filas):")
print(mask_nan.head())

# Devuelve True si hay al menos un valor ausente
print("\n¿Hay al menos un valor ausente?:", mask_nan.any().any())

# Devuelve el número total de NaN que presenta el DataFrame
total_nan = mask_nan.sum().sum()
print(f"\nTotal de NaN en el DataFrame: {total_nan}")

# Devuelve el % de valores perdidos
total_celdas = calidad_aire.size
pct_total = total_nan / total_celdas * 100
print(f"% total de valores perdidos: {pct_total:.2f}%")

# Detección del número de valores perdidos en cada una de las columnas
nan_por_columna = calidad_aire.isnull().sum().sort_values(ascending=False)
print("\nValores perdidos por columna:")
print(nan_por_columna)

# Detección del % de valores perdidos en cada una de las columnas
pct_por_columna = (calidad_aire.isnull().sum() / len(calidad_aire) * 100).sort_values(ascending=False)
print("\n% de valores perdidos por columna:")
print(pct_por_columna.round(2))
```

### Tratamiento de datos ausentes (eliminación e imputación por la media)

```python
# Guardar dataset original
calidad_aire_original = calidad_aire.copy()

# Eliminación de las variables que presentan un % de NaN superior al 50%
pct_na = calidad_aire.isnull().mean() * 100
columnas_a_eliminar = pct_na[pct_na > 50].index.tolist()
print("Columnas eliminadas por >50% de NaN:", columnas_a_eliminar)

calidad_aire_imputado = calidad_aire.drop(columns=columnas_a_eliminar)

# Seleccionamos las variables numéricas
variables_numericas = calidad_aire_imputado.select_dtypes(include=np.number).columns.tolist()
print("Variables numéricas para imputar:", variables_numericas)

# Calculamos la media para cada una de las variables numéricas sin tener en cuenta los NaN
medias = calidad_aire_imputado[variables_numericas].mean()
print("\nMedias por variable:")
print(medias.round(3))

# Sustituimos los valores NaN por la media correspondiente a cada variable
calidad_aire_imputado[variables_numericas] = calidad_aire_imputado[variables_numericas].fillna(medias)

# Verificación
print("\nNaN restantes por columna después de imputar:")
print(calidad_aire_imputado.isnull().sum())
```

## 4. **Detección de datos atípicos en variablers numéricas (Histograma, gráfico de cajas y gráfico de bigotes)**

```python
# Histograma para analizar la distribución de ozono (O3)
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

if 'O3 (ug/m3)' in calidad_aire_imputado.columns:
    serie_o3 = calidad_aire_imputado['O3 (ug/m3)'].dropna()
    axes[0].hist(serie_o3, bins=60, color='teal', edgecolor='black')
    axes[0].set_title('Histograma de O3 (ug/m3)')
    axes[0].set_xlabel('O3 (ug/m3)')
    axes[0].set_ylabel('Frecuencia')

    # Estadísticas necesarias para reproducir el gráfico de cajas y bigotes
    q1 = serie_o3.quantile(0.25)
    q2 = serie_o3.quantile(0.50)
    q3 = serie_o3.quantile(0.75)
    iqr = q3 - q1
    bigote_inf = max(serie_o3.min(), q1 - 1.5 * iqr)
    bigote_sup = min(serie_o3.max(), q3 + 1.5 * iqr)

    print("Estadísticas para el boxplot de O3:")
    print(f"  Q1 = {q1:.2f}")
    print(f"  Mediana (Q2) = {q2:.2f}")
    print(f"  Q3 = {q3:.2f}")
    print(f"  IQR = {iqr:.2f}")
    print(f"  Bigote inferior = {bigote_inf:.2f}")
    print(f"  Bigote superior = {bigote_sup:.2f}")

    # Construcción del gráfico de cajas y bigotes
    axes[1].boxplot(serie_o3, vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightblue'))
    axes[1].set_title('Boxplot de O3 (ug/m3)')
    axes[1].set_ylabel('O3 (ug/m3)')

plt.tight_layout()
plt.show()
```

### Detección de valores atípicos en variables categóricas

```python
# Número de categorías que presenta la variable Provincia
conteo_provincias = calidad_aire_imputado['Provincia'].value_counts()
print(f"Número de categorías en Provincia: {conteo_provincias.shape[0]}")
print("\nFrecuencia por provincia:")
print(conteo_provincias)

# Construcción del gráfico de barras para la variable Provincia
fig, ax = plt.subplots(figsize=(10, 5))
conteo_provincias.plot(kind='bar', color='coral', edgecolor='black', ax=ax)
ax.set_title('Número de registros por Provincia')
ax.set_xlabel('Provincia')
ax.set_ylabel('Número de registros')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Tratamiento de valores atípicos en variables numéricas

```python
# Se genera una nueva tabla que no contiene los valores identificados como atípicos
# Aplicamos la regla del rango intercuartílico (IQR) sobre las variables de contaminantes
variables_contaminantes = [c for c in calidad_aire_imputado.select_dtypes(include=np.number).columns
                           if c not in ('Latitud', 'Longitud')]

calidad_aire_sin_outliers = calidad_aire_imputado.copy()
for col in variables_contaminantes:
    q1 = calidad_aire_sin_outliers[col].quantile(0.25)
    q3 = calidad_aire_sin_outliers[col].quantile(0.75)
    iqr = q3 - q1
    limite_inf = q1 - 1.5 * iqr
    limite_sup = q3 + 1.5 * iqr
    antes = len(calidad_aire_sin_outliers)
    calidad_aire_sin_outliers = calidad_aire_sin_outliers[
        (calidad_aire_sin_outliers[col] >= limite_inf) &
        (calidad_aire_sin_outliers[col] <= limite_sup)
    ]
    despues = len(calidad_aire_sin_outliers)
    print(f"{col}: eliminadas {antes - despues} filas atípicas (límites: {limite_inf:.2f}, {limite_sup:.2f})")

print(f"\nTamaño antes: {len(calidad_aire_imputado)}, después: {len(calidad_aire_sin_outliers)}")

# Construcción de los gráficos de cajas y bigotes (antes vs después)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
calidad_aire_imputado[variables_contaminantes].boxplot(ax=axes[0])
axes[0].set_title('Antes del tratamiento de outliers')
axes[0].tick_params(axis='x', rotation=45)

calidad_aire_sin_outliers[variables_contaminantes].boxplot(ax=axes[1])
axes[1].set_title('Después del tratamiento de outliers')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Tratamiento de valores atípicos en variables categóricas

```python
# Identificamos las provincias con muy pocos registros (categorías atípicas)
conteo = calidad_aire_sin_outliers['Provincia'].value_counts()
print("Frecuencia actual por provincia:")
print(conteo)

# Si existe "Madrid" la eliminamos; si no, eliminamos las provincias con frecuencia muy baja
provincias_a_eliminar = []
if 'Madrid' in conteo.index:
    provincias_a_eliminar.append('Madrid')

# Adicionalmente, eliminamos provincias con menos del 1% del total (categorías raras)
umbral = len(calidad_aire_sin_outliers) * 0.01
raras = conteo[conteo < umbral].index.tolist()
provincias_a_eliminar = list(set(provincias_a_eliminar + raras))

print(f"\nProvincias a eliminar (atípicas): {provincias_a_eliminar}")

# Creamos una copia explícita del DataFrame excluyendo esas provincias
calidad_aire_final = calidad_aire_sin_outliers[
    ~calidad_aire_sin_outliers['Provincia'].isin(provincias_a_eliminar)
].copy()

# Eliminamos las categorías no utilizadas
calidad_aire_final['Provincia'] = calidad_aire_final['Provincia'].cat.remove_unused_categories()
calidad_aire_final['Estación'] = calidad_aire_final['Estación'].cat.remove_unused_categories()

# Verificamos la eliminación
print("\nProvincias finales:")
print(calidad_aire_final['Provincia'].value_counts())
print(f"\nTamaño final del DataFrame: {calidad_aire_final.shape}")
```

## 5. **Análisis de correlación de variables**

```python
import seaborn as sns

# Calculamos la matriz de coeficientes de correlación entre las variables numéricas
variables_corr = [c for c in calidad_aire_final.select_dtypes(include=np.number).columns
                  if c not in ('Latitud', 'Longitud')]
matriz_corr = calidad_aire_final[variables_corr].corr(method='pearson')
print("Matriz de correlación de Pearson:")
print(matriz_corr.round(2))

# Configuración del gráfico de correlación
plt.figure(figsize=(8, 6))

# Gráfico de correlaciones utilizando un mapa de calor
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt='.2f',
            linewidths=0.5, cbar_kws={'label': 'Coeficiente de Pearson'})
plt.title('Matriz de correlación entre contaminantes')
plt.tight_layout()
plt.show()
```

# **Ejercicios**
1. Realice un EDA con el dataset de Data set de Ordenes de Compra y ordenes de Servicio años 2025 [MPSI] de la plataforma Nacional de Datos Abiertos del Perú (link: https://datosabiertos.gob.pe/dataset/%C3%B3rdenes-de-compra-y-ordenes-de-servicio-a%C3%B1os-2025-municipalidad-provincial-san-ignacio-0#)


```python
# Ejercicio 1: EDA con un dataset de Órdenes de Compra y Servicio (MPSI - 2025).
# Se intenta descargar el CSV real desde la Plataforma Nacional de Datos Abiertos del Perú.
# Si no hay conexión o el endpoint no responde, se genera un dataset sintético equivalente
# para demostrar el flujo completo de EDA.

import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

URL_MPSI = ("https://datosabiertos.gob.pe/sites/default/files/"
            "ordenes_compra_servicio_mpsi_2025.csv")

def cargar_mpsi():
    try:
        return pd.read_csv(URL_MPSI, sep=';', encoding='latin-1')
    except Exception as e:
        print(f"No se pudo descargar el CSV oficial ({e}).")
        print("Generando dataset sintético equivalente para demostración...")
        rng = np.random.default_rng(42)
        n = 500
        productos = ['Materiales de oficina', 'Combustible', 'Equipos informáticos',
                     'Servicios de limpieza', 'Mantenimiento vehicular', 'Insumos médicos',
                     'Alimentos', 'Mobiliario']
        proveedores = [f'Proveedor {chr(65 + i)}' for i in range(15)]
        modalidades = ['Adjudicación directa', 'Selección abreviada', 'Concurso público',
                       'Compra menor']
        df = pd.DataFrame({
            'numero_orden': range(1, n + 1),
            'fecha': pd.date_range('2025-01-01', periods=n, freq='D').to_series().sample(n, replace=True, random_state=42).reset_index(drop=True),
            'tipo': rng.choice(['Compra', 'Servicio'], size=n, p=[0.6, 0.4]),
            'modalidad': rng.choice(modalidades, size=n),
            'producto_servicio': rng.choice(productos, size=n),
            'proveedor': rng.choice(proveedores, size=n),
            'monto_soles': np.round(rng.lognormal(mean=8, sigma=1.2, size=n), 2),
            'cantidad': rng.integers(1, 50, size=n),
        })
        # Inyectamos algunos NaN para que la limpieza tenga sentido
        df.loc[rng.choice(n, size=20, replace=False), 'monto_soles'] = np.nan
        df.loc[rng.choice(n, size=15, replace=False), 'proveedor'] = np.nan
        return df

mpsi = cargar_mpsi()

# a) Descripción inicial
print("Forma:", mpsi.shape)
print("\nPrimeras filas:")
print(mpsi.head())
print("\nTipos de datos:")
print(mpsi.dtypes)

# b) Limpieza
print("\nValores faltantes:")
print(mpsi.isnull().sum())

mpsi_clean = mpsi.copy()
# Imputamos monto con la mediana y proveedor con 'Desconocido'
mpsi_clean['monto_soles'] = mpsi_clean['monto_soles'].fillna(mpsi_clean['monto_soles'].median())
if 'proveedor' in mpsi_clean.columns:
    mpsi_clean['proveedor'] = mpsi_clean['proveedor'].fillna('Desconocido')

# c) Estadísticas descriptivas
print("\nResumen numérico:")
print(mpsi_clean.describe())

# d) Análisis de relaciones — top proveedores y por modalidad
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
top_prov = mpsi_clean.groupby('proveedor')['monto_soles'].sum().nlargest(10)
top_prov.plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Top 10 proveedores por monto total (S/.)')
axes[0].set_xlabel('Monto total')

if 'modalidad' in mpsi_clean.columns:
    sns.boxplot(data=mpsi_clean, x='modalidad', y='monto_soles', ax=axes[1])
    axes[1].set_title('Distribución de montos por modalidad')
    axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()

# Distribución del monto
plt.figure(figsize=(10, 4))
sns.histplot(mpsi_clean['monto_soles'], bins=40, kde=True, color='darkorange')
plt.title('Distribución del monto de las órdenes (S/.)')
plt.xlabel('Monto (S/.)')
plt.show()
```

### Conclusiones del EDA - MPSI

1. **Distribución de montos:** la mayoría de las órdenes se concentran en montos bajos a medios y existe una cola larga de pocas órdenes con montos elevados (distribución asimétrica), típica de gasto público.

```python
# Ejercicio 2: Comparativa de herramientas de EDA-automático.
# Probaremos Sweetviz y D-Tale (alternativa: AutoViz). Si no están instaladas,
# generamos una comparativa documentada y un mini-reporte manual con seaborn.

import sys, subprocess
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Usamos una muestra del dataset de calidad del aire para que sea rápido
muestra_eda = calidad_aire_final.sample(n=min(5000, len(calidad_aire_final)), random_state=0).copy()

# ---------- Sweetviz ----------
sweetviz_ok = False
try:
    import sweetviz as sv
    sweetviz_ok = True
except ImportError:
    print("Sweetviz no está instalado. Instalación opcional: pip install sweetviz")

if sweetviz_ok:
    reporte_sv = sv.analyze(muestra_eda, pairwise_analysis='off')
    reporte_sv.show_html('sweetviz_calidad_aire.html', open_browser=False)
    print("Sweetviz: reporte generado en sweetviz_calidad_aire.html")
else:
    # Mini reporte manual estilo Sweetviz (resumen + histogramas)
    print("\n[Fallback Sweetviz] Mini-resumen del dataset:")
    print(muestra_eda.describe(include='all'))
    fig, axes = plt.subplots(2, 3, figsize=(14, 8))
    cols_num = muestra_eda.select_dtypes(include=np.number).columns[:6]
    for ax, col in zip(axes.flatten(), cols_num):
        sns.histplot(muestra_eda[col].dropna(), ax=ax, kde=True, color='teal')
        ax.set_title(col)
    plt.tight_layout()
    plt.show()

# ---------- AutoViz ----------
autoviz_ok = False
try:
    from autoviz.AutoViz_Class import AutoViz_Class
    autoviz_ok = True
except ImportError:
    print("\nAutoViz no está instalado. Instalación opcional: pip install autoviz")

if autoviz_ok:
    av = AutoViz_Class()
    av.AutoViz(filename='', dfte=muestra_eda, depVar='', verbose=0,
               max_rows_analyzed=5000, max_cols_analyzed=15, chart_format='png')
    print("AutoViz: gráficos generados arriba.")
else:
    # Mini reporte manual estilo AutoViz (matriz de correlación + scatter)
    print("\n[Fallback AutoViz] Matriz de correlación de la muestra:")
    cols_num = muestra_eda.select_dtypes(include=np.number).columns
    plt.figure(figsize=(8, 6))
    sns.heatmap(muestra_eda[cols_num].corr(), annot=True, cmap='vlag', fmt='.2f')
    plt.title('Correlaciones (fallback AutoViz)')
    plt.tight_layout()
    plt.show()

print("\nResumen:")
print(f"  Sweetviz disponible: {sweetviz_ok}")
print(f"  AutoViz  disponible: {autoviz_ok}")
```

### Comparativa Sweetviz vs AutoViz

| Aspecto | Sweetviz | AutoViz |
|---|---|---|
| Tipo de salida | Reporte HTML interactivo (auto-contained) | Series de gráficos PNG/Jupyter inline |
| Comparación entre datasets | Sí, soporta `compare()` (train vs test) | No es su foco principal |
| Análisis bivariado automático | Asocia cada feature con el target si se especifica | Detecta tipos y elige gráficos según cardinalidad |
| Velocidad en datasets grandes | Buena con `pairwise_analysis='off'` | Muy buena, muestrea por defecto |
| Curva de aprendizaje | Muy baja, una sola llamada | Baja, configuración por parámetros |
| Dependencias | Pocas (matplotlib, pandas) | Más pesadas (xgboost, hdbscan opcional) |

**Conclusión:** Sweetviz es más adecuado cuando necesitas un *informe distribuible* (HTML único) y comparar conjuntos de datos. AutoViz brilla cuando quieres descubrir rápidamente las visualizaciones más informativas en un notebook, sin generar un archivo aparte. Para un primer vistazo en clase, Sweetviz es más cómodo; para exploración iterativa, AutoViz aporta variedad de gráficos sin escribir código.
