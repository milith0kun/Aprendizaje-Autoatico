#**CODIFICACIÓN DE DATOS CATEGÓRICOS**

Las datos categóricos son aquellos que contienen valores discretos y no numéricos, como denominaciones, nombres de ciudades, géneros, colores, codificación textual, etc.<br>
Los algoritmos de machine learning, especialmente los basados en álgebra lineal, no pueden trabajar directamente con valores categóricos, por lo que éstos deben ser convertidos a números.<br>
La codificación de datos categóricos en el preprocesamiento de datos se refiere a la transformación de variables categóricas en un formato que pueda ser proporcionado a los algoritmos de machine learning para mejorar el rendimiento y precisión del modelo.<br>

##Generación del dataset

```python
import pandas as pd

# -- Definir una función que devuelva un DataFrame de ejemplo
def DatasetEjemplo():
  data = {'Animales': ['Perro', 'Gato', 'Perro', 'Ave'], 'Frutas': ['Manzana', 'Plátano', 'Pera', 'Fresa']}
  return pd.DataFrame(data)
# -- Mostrar el dataset de ejemplo
dataset = DatasetEjemplo()
print(dataset)
print(dataset.dtypes)
```

##Algoritmos de codificación de datos categóricos

###**1.- Utilizando diccionarios**

####Base teórica

La codificación de valores categóricos utilizando diccionarios implica crear un diccionario donde las claves son las categorías y los valores son los códigos numéricos correspondientes. Este diccionario se usa para transformar los valores categóricos en números que pueden ser utilizados por los algoritmos de machine learning.<br>
Dentro de las ventajas de usar Diccionarios para la codificación de valores categóricos, se tiene: Simplicidad y Claridad, Flexibilidad, Eficiencia, Facilidad de Reutilización y Consistencia.<br>
Los casos en los que se recomienda usar diccionarios para la codificación de valores categóricos son:<br>
*   ***Pequeño Número de Categorías***: Esta técnicaes particularmente útil cuando el número de categorías es pequeño y conocido de antemano.
*   ***Necesidad de Control sobre la Codificación***: Si necesitas asignar valores específicos a las categorías (por ejemplo, para mantener un orden lógico), los diccionarios te permiten hacerlo de manera sencilla.
*   ***Datos con Categorías Recurrentes***: Si estás trabajando con datos donde las mismas categorías se repiten en diferentes columnas o conjuntos de datos, un diccionario permite mantener la consistencia en la codificación.
*   ***Preprocesamiento Personalizado***: En situaciones donde los métodos automáticos de codificación (como Label Encoding o One-Hot Encoding) no se ajustan a tus necesidades específicas, crear tu propio diccionario te da la flexibilidad necesaria.

####Ejemplo

```python
# Recuperar dataset
dataset = DatasetEjemplo()
# Diccionario de reemplazo
replace_dict = {'Animales': {'Perro': 1, 'Gato': 2, 'Ave': 3},
                'Frutas': {'Manzana': 1, 'Plátano': 2, 'Pera': 3, 'Fresa':4}}

# Reemplazar los valores categóricos con numéricos
dataset.replace(replace_dict, inplace=True)
print(dataset)
print(dataset.dtypes)
```

###**2.-Utilizando Librería** ***LabelEncoder***

####Base teórica

LabelEncoder es una clase proporcionada por la biblioteca scikit-learn en Python que se utiliza para convertir datos categóricos en valores numéricos. Esta técnica de codificación es simple y eficiente, especialmente cuando se tiene un número limitado de categorías en los datos. ***LabelEncoder*** asigna un número entero único a cada categoría en los datos.<br>
Las ventajas de usar ***LabelEncoder*** son:
Simplicidad y Eficiencia, Consistencia y
Compatibilidad.<br>
Los casos en los que se recomienda usar ***LabelEncoder*** son:
*   ***Variables Ordinales***: Si tus categorías tienen un orden intrínseco (como calificaciones: 'malo', 'regular', 'bueno'), ***LabelEncoder*** puede ser adecuado porque mantiene el orden.
*   ***Modelos específicos***: Algunos algoritmos como los árboles de decisión y los algoritmos basados en ensamblaje (ensemble methods) pueden manejar adecuadamente las variables categóricas codificadas como enteros.
*   ***Datos con pocas categorías***: Cuando tienes un conjunto de datos con un número limitado de categorías, ***LabelEncoder*** es simple y efectivo.

####Ejemplo

```python
# Definir la libreria
from sklearn.preprocessing import LabelEncoder
# Definir el objeto LabelEncoder
le = LabelEncoder()
# Recuperar dataset
dataset = DatasetEjemplo()
# Procesar cada columna categórica
for column in dataset.select_dtypes(include=['object']).columns:
    dataset[column] = le.fit_transform(dataset[column])
# Mostrar dataset
print(dataset)
print(dataset.dtypes)
```

###**3.- Utilizando Librería** ***OrdinalEncoder***

####Base teórica

***OrdinalEncoder*** es una clase proporcionada por la biblioteca scikit-learn en Python que se utiliza para convertir variables categóricas en valores numéricos, manteniendo el orden intrínseco de las categorías. Esta técnica es adecuada para variables ordinales, donde las categorías tienen un orden natural. ***OrdinalEncoder*** asigna números enteros a cada categoría, respetando el orden de las mismas.<br>
Las ventajas de utilizar ***OrdinalEncoder*** son: Mantiene el Orden, Simplicidad, Eficiencia y Compatibilidad.<br>
Los casos en los que se Recomienda Usar ***OrdinalEncoder*** son:
*   ***Variables Ordinales***: Cuando las categorías tienen un orden intrínseco, como calificaciones (malo, regular, bueno), niveles de satisfacción (bajo, medio, alto), etc.
*   ***Modelos que Toman en Cuenta el Orden****: Algoritmos como los árboles de decisión y los modelos basados en ensamblaje (ensemble methods) pueden manejar adecuadamente las variables ordinales codificadas.
*   ***Datos con Pocas Categorías***: Es útil cuando tienes un número limitado de categorías con un orden definido.








####Ejemplo

```python
# Definir la libreria
from sklearn.preprocessing import OrdinalEncoder
# Definir el objeto OrdinalEncoder
oe = OrdinalEncoder()
# Recuperar dataset
dataset = DatasetEjemplo()

# Efectuar el proceso de transformación
dataset[['Animales', 'Frutas']] = oe.fit_transform(dataset[['Animales', 'Frutas']])

print(dataset)
print(dataset.dtypes)
```

###**4.- Utilizando** ***cat.codes***

####Base teórica

El atributo cat.codes se utiliza en pandas para convertir valores categóricos en códigos numéricos. Es una manera conveniente y eficiente de realizar codificación de variables categóricas directamente en un DataFrame. ***cat.codes*** es un atributo de las columnas categóricas en pandas. Cuando se convierte una columna a un tipo de datos categóricos, pandas asigna internamente códigos numéricos a cada categoría única. ***cat.codes*** expone estos códigos como una serie de pandas, donde cada categoría se mapea a un número entero.<br>
Las ventajas de utilizar ***cat.codes*** son: Simplicidad, Eficiencia, Integración y Flexibilidad. <br>
Casos en los que se Recomienda Usar ***cat.codes*** son:
*   ***Análisis exploratorio de datos*** (EDA): Durante la fase de exploración de datos, cuando deseas rápidamente convertir categorías a números para análisis o visualización.
*   ***Preparación de datos para modelado***: Antes de alimentar datos a algoritmos de machine learning que requieren entradas numéricas.
*   ***Variables Ordinales y Nominales***: Es adecuado tanto para variables ordinales (donde el orden importa) como para variables nominales (donde el orden no importa), aunque se debe tener cuidado con el significado de los números asignados en el caso de variables nominales.

####Ejemplo

```python
# Recuperar dataset
dataset = DatasetEjemplo()
# Codificación usando cat.codes
dataset['Animales'] = dataset['Animales'].astype('category').cat.codes
dataset['Frutas'] = dataset['Frutas'].astype('category').cat.codes

print(dataset)
print(dataset.dtypes)
```
